# Visualize My Math

```python
# Initialization
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

### Part 1: *Funcions*

$$
y=x^2
$$

```python
x = np.linspace(-10, 10, 20)
y = x**2

plt.plot(x, y)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
ax = plt.gca()
ax.set_aspect(1.0)
```

$$
y=\frac{1}{x}
$$

```python
x = np.linspace(-10, 10, 90)

plt.plot(x, 1/x)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
ax = plt.gca()
ax.set_aspect(1.0)
```

$$
y=\sqrt{x}
$$

```python
x = np.linspace(0, 10, 100)

plt.plot(x, x**(1/2))
plt.axhline(0, color='black')
plt.axvline(0, color='black')
ax = plt.gca()
ax.set_aspect(1.0)
```

$$
y=\sqrt{4-x}
$$

```python
x = np.linspace(0, 10, 100)

plt.plot(x, (4-x)**(1/2))
plt.axhline(0, color='black')
plt.axvline(0, color='black')
ax = plt.gca()
ax.set_aspect(1.0)
```

$$
y=sin(x)
$$

```python
x = np.linspace(-3*np.pi, 3*np.pi, 100)

plt.plot(x, np.sin(x))
plt.axhline(0, color='black')
plt.axvline(0, color='black')
ax = plt.gca()
ax.set_aspect(1.0)
```
