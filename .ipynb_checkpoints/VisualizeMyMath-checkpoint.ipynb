# Visualize My Math

```python
# Initialization
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

### Part 1: *Funcions*

\begin{equation}
y=x^2
\end{equation}

```python


x = np.linspace(-3*np.pi, 3*np.pi, 100)
y = np.sin(x)



graph = sns.lineplot(x=x, y=y)

plt.axhline(0, color='k')
plt.axvline(0, color='k')
```

\begin{equation}
y=sin(x)
\end{equation}

```python
x = np.linespace(-2*np.pi, 2*np.pi, 500)
y = np.sin(x)

sns.set_theme()

graph = sns.lineplot(x=x, y=y)

plt.axhline(0, color='k')
plt.axvline(0, color='k')
```
